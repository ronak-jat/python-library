# bar graph

```python
x=['python','java','c','c++','javascript','c#','php']
y=[85,80,60,54,77,45,30]
z=[60,45,21,32,54,25,10]
plt.xlabel("languages",fontsize=15)
plt.ylabel("percentage",fontsize=15)
plt.title("survey of languages used",fontsize=15)
c=['g','b','y','m','g','m','b']
plt.bar(x,y,width=0.4,color='b',edgecolor='black',linewidth=2,alpha=0.5,label='popularity1')
plt.bar(x,z,width=0.4,color='m',alpha=0.5,label='popularity2')
plt.legend()
plt.show()
# aligh is used to give edge to x axis name
#edgecolor is used to give color to edges of bar and line width is for border
#alpha is used to light or dull the color
# linestyle is used to give style to border
# label is used to label use it with legend
```

```python
x=['python','java','c','c++','javascript','c#','php']
y=[85,80,60,54,77,45,30]
z=[60,45,21,32,54,25,10]

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]
# if we add width with index then the second bar graph will shift to right

plt.xlabel("languages",fontsize=15)
plt.ylabel("percentage",fontsize=15)
plt.title("survey of languages used",fontsize=15)

plt.bar(p,y,width,color='b',edgecolor='black',linewidth=2,alpha=0.5,label='popularity1')
plt.bar(p1,z,width,color='m',edgecolor='black',linewidth=2,alpha=0.5,label='popularity2')
plt.xticks(p+width/2,x,rotation=10)
# p+width for position 
# rotaton is used to rotate x axis elements name and xticks to provide that name
plt.legend()
plt.show()
```

```python
x=['python','java','c','c++','javascript','c#','php']
y=[85,80,60,54,77,45,30]
z=[60,45,21,32,54,25,10]

plt.xlabel("percentage",fontsize=15)
plt.ylabel("languages",fontsize=15)
plt.title("survey of languages used",fontsize=15)

plt.barh(x,y,color='b',alpha=0.5,label='popularity1')
plt.barh(x,z,color='m',alpha=0.5,label='popularity2')
#barh is used for horizontal bar graph and width can't be used in it
plt.legend()
plt.show()
```

# pie chart

```python
x=['python','c','html','css','javascript']
y=[45,7,15,19,14]
ex=[0.0,0.0,0.0,0.0,0.2]
plt.pie(y,labels=x,explode=ex,autopct='%0.1f%%',shadow=True,radius=1,labeldistance=1.2,startangle=352,
       textprops={"fontsize":11},counterclock=False,wedgeprops={"linewidth":6})
plt.title("languages i know",fontsize=15)
plt.legend(loc=3)
plt.show()
# label is used to give label to chart
# explode is used to take out the pie out of chart
#autopct is used to give percentage to graph 
# shadow is used to give shadow to graph basically to make it pretty
# radius is used to change the radius of circle basically size
# labeldistance decide the distance between label and graph 
#startangle is used to change the starting angle off graph
# you can use color to ,colors=c c=['m','g','b','y','r']
# textprops are used to change the size of label
# counterclock is used to change direction of piechart
# wedgeprops are used to change the width of chart
# "edgecolor":"black" you can give it too 
# rotatelabels=True you can use it to rotate label
# The center parameter in a pie chart allows you to control the position of the pie chart within the plot.
#,center=(5,2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for monthly sales
months = pd.date_range('2023-01-01', '2023-12-01', freq='MS')
sales = [120, 150, 160, 180, 200, 210, 240, 250, 270, 300, 320, 330]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(months, sales, marker='o', color='b', label='Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data for 2023')
plt.grid(True)
plt.legend()

# Highlight a specific period
plt.axvspan('2023-06-01', '2023-08-01', color='gray', alpha=0.2, label='Summer Sale Period')
plt.legend()
plt.show()
```

```python
import matplotlib.dates as mdates

# Sample stock price data
dates = pd.date_range('2023-01-01', '2023-12-31', freq='W')
prices = np.cumsum(np.random.randn(len(dates)) * 5 + 100)  # Simulated stock prices

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(dates, prices, color='darkgreen', label='Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price Over Time')
plt.grid(True)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Major ticks every month
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))

# Highlight high and low points
plt.scatter(dates[np.argmax(prices)], max(prices), color='red', label='Highest Price')
plt.scatter(dates[np.argmin(prices)], min(prices), color='blue', label='Lowest Price')
plt.legend()
plt.show()
```

```python
# Data for stacked area plot
x = np.arange(0, 10, 0.1)
y1 = np.sin(x) + 1
y2 = np.cos(x) + 1
y3 = np.sin(x / 2) + 0.5

# Create stacked area plot
plt.fill_between(x, y1, color="skyblue", alpha=0.5, label="Series 1")
plt.fill_between(x, y2 + y1, y1, color="salmon", alpha=0.5, label="Series 2")
plt.fill_between(x, y3 + y2 + y1, y2 + y1, color="lightgreen", alpha=0.5, label="Series 3")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Stacked Area Plot")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Create area plot
plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.plot(x, y, color="Slateblue", alpha=0.7)  # Line for better readability
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Area Plot")
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Highlight only positive regions of y
plt.fill_between(x, y, where=(y > 0), color="skyblue", alpha=0.5, label="Positive Region")
plt.fill_between(x, y, where=(y < 0), color="salmon", alpha=0.5, label="Negative Region")
plt.plot(x, y, color="Slateblue")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Conditional Area Plot")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Add transparency to overlapping area plots
plt.fill_between(x, y1, color="blue", alpha=0.3, label="sin(x)")
plt.fill_between(x, y2, color="red", alpha=0.3, label="cos(x)")
plt.plot(x, y1, color="blue", alpha=0.6)
plt.plot(x, y2, color="red", alpha=0.6)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Area Plot with Transparency")
plt.legend()
plt.show()
```
