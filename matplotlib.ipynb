```python
x=['python','java','c','c++','javascript','c#','php']
y=[85,80,60,54,77,45,30]
z=[60,45,21,32,54,25,10]

width=0.3
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=15)
plt.ylabel("percentage",fontsize=15)
plt.title("survey of languages used",fontsize=15)

plt.bar(p,y,width,color='b',edgecolor='black',linewidth=2,alpha=0.5,label='popularity1')
plt.bar(p1,z,width,color='m',edgecolor='black',linewidth=2,alpha=0.5,label='popularity2')
plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

months = pd.date_range('2023-01-01', '2023-12-01', freq='MS')
sales = [120, 150, 160, 180, 200, 210, 240, 250, 270, 300, 320, 330]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(months, sales, marker='o', color='b', label='Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data for 2023')
plt.grid(True)
plt.legend()

plt.axvspan('2023-06-01', '2023-08-01', color='gray', alpha=0.2, label='Summer Sale Period')
plt.legend()
plt.show()
```

```python
categories = ['Electronics', 'Furniture', 'Clothing', 'Books']
sales_Q1 = [200, 150, 400, 300]
sales_Q2 = [250, 160, 420, 310]
width=0.3
p=np.arange(4)
p1=[i+width for i in p]

plt.figure(figsize=(10, 6))
plt.xlabel('Product Categories')
plt.ylabel('Sales')
plt.title('Sales Comparison by Product Category')

plt.bar(p,sales_Q1,color='teal',width=0.3,label='Q1')
plt.bar(p1,sales_Q2,color='orange',width=0.3,label='Q2')
plt.xticks(p+width/2,categories)

plt.legend()
plt.show()
```

```python
import matplotlib.dates as mdates

dates = pd.date_range('2023-01-01', '2023-12-31', freq='W')
prices = np.cumsum(np.random.randn(len(dates)) * 5 + 100)  

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(dates, prices, color='darkgreen', label='Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price Over Time')
plt.grid(True)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator()) 
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))

plt.scatter(dates[np.argmax(prices)], max(prices), color='red', label='Highest Price')
plt.scatter(dates[np.argmin(prices)], min(prices), color='blue', label='Lowest Price')
plt.legend()
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y1 = np.sin(x) + 1
y2 = np.cos(x) + 1
y3 = np.sin(x / 2) + 0.5

plt.fill_between(x, y1, color="skyblue", alpha=0.5, label="Series 1")
plt.fill_between(x, y2 + y1, y1, color="salmon", alpha=0.5, label="Series 2")
plt.fill_between(x, y3 + y2 + y1, y2 + y1, color="lightgreen", alpha=0.5, label="Series 3")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Stacked Area Plot")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Create area plot
plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.plot(x, y, color="Slateblue", alpha=0.7)  
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Area Plot")
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.fill_between(x, y, where=(y > 0), color="skyblue", alpha=0.5, label="Positive Region")
plt.fill_between(x, y, where=(y < 0), color="salmon", alpha=0.5, label="Negative Region")
plt.plot(x, y, color="Slateblue")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Conditional Area Plot")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.fill_between(x, y1, color="blue", alpha=0.3, label="sin(x)")
plt.fill_between(x, y2, color="red", alpha=0.3, label="cos(x)")
plt.plot(x, y1, color="blue", alpha=0.6)
plt.plot(x, y2, color="red", alpha=0.6)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Area Plot with Transparency")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 10, 1)
y = np.random.randint(1, 10, size=10)

fig=plt.figure(figsize=(10,5))
plt.fill_between(x, y, color="skyblue", step='mid', alpha=0.5)
plt.plot(x, y, color="Slateblue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Stepwise Area Plot")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + 1
y2 = np.cos(x) + 1.5

plt.fill_between(x, y1, color="blue", alpha=0.3, label="sin(x) + 1")
plt.fill_between(x, y2, color="orange", alpha=0.3, label="cos(x) + 1.5")

plt.plot(x, y1, color="blue", alpha=0.6)
plt.plot(x, y2, color="orange", alpha=0.6)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Advanced Area Plot")
plt.legend()
plt.grid(True)
plt.show()
```
