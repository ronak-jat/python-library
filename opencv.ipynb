```python
img=cv2.imread("pikachu.jpg",1)
edge= cv2.Canny(img,125,175) # to find edges of image

erosion = cv2.erode(edge,(3,3),iterations = 2)  # to decrease the pixel or width of the edges

dilation = cv2.dilate(edge,(3,3),iterations = 1)  # to increase the pixel or width of the edges # opposite of erosion

cv2.imshow('Original', img)
cv2.imshow('Erosion', erosion)
cv2.imshow('Dialation', dilation)
cv2.imshow("canny edges",edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to draw circle on image when click

img = cv2.imread('pikachu.jpg')

def draw(events,x,y,flags,params):
    if events==1:
        cv2.circle(img,center=(x,y),radius=30,color=(233,0,0),thickness=-1)

cv2.namedWindow("window")
cv2.setMouseCallback("window",draw)

while True:
    cv2.imshow("window",img)
    if cv2.waitKey(1) & 0xff == ord("v"): # press v to close window
        break
cv2.destroyAllWindows()
```

```python
# to draw rectangle on drag and drop

img = cv2.imread('pikachu.jpg')
drawing=False
def draw(events,x,y,flags,params):
    global fx,fy
    if events==1:
        drawing=True
        fx=x
        fy=y
    elif events==4:
        drawing=False
        cv2.rectangle(img,pt1=(fx,fy),pt2=(x,y),color=(255,0,0),thickness=-1)

cv2.namedWindow("window")
cv2.setMouseCallback("window",draw)

while True:
    cv2.imshow("window",img)
    if cv2.waitKey(1) & 0xff == ord("v"):
        break
cv2.destroyAllWindows()
```

```python
drawing=True
shape='r'

def draw_circle(event,x,y,flags,param):
    global x1,x2
    if event == 1:
        drawing = True
        x1,x2 = x,y
    elif event == 4:
        drawing = False
        if shape == 'r':
            cv2.rectangle(img,(x1,x2),(x,y),(0,255,0),-1)
        if shape == 'l':
            cv2.line(img,(x1,x2),(x,y),(255,255,255),3)
        if shape=='c':
            cv2.circle(img,(x,y), 10, (255,255,0), -1)
            
img = cv2.imread('pikachu.jpg')
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)
while(1):
    cv2.imshow('image',img)
    key=cv2.waitKey(1)
    if key==ord('1'):
        shape='r'
    if key==ord('2'):
        shape='l'
    if key==ord('3'):
        shape='c'
#print (shape)
    if key == ord("v"):
        break
cv2.destroyAllWindows()
```

```python
img = cv2.imread('pikachu.jpg')
def translate(img, x, y):
    transMat = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1], img.shape[0])
    return cv2.warpAffine(img, transMat, dimensions)

# -x --> Left
# -y --> Up
# x --> Right
# y --> Down

translated = translate(img, 60, 70)
cv2.imshow('Translated', translated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('chill sloth.png')
blank=np.zeros(image.shape)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
cv2.drawContours(blank, contours, -1, (0, 255, 0), 1)

plt.imshow(blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Grayscale histogram of image 
image = cv2.imread('pikachu.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hist= cv2.calcHist([gray],[0],None, [255],[0,255])
plt.plot(hist) 
```

```python
#Plotting color histogram
image = cv2.imread('chill sloth.png')
blank= np.zeros(image.shape[:2], dtype=np.uint8)
blank= cv2.circle(blank,(400,500),200,(255,255,255),-1)
mask=cv2.bitwise_and(image,image,mask=blank)
colors=("b","g","r")
for i,col in enumerate(colors):
    hist= cv2.calcHist([image],[i],None, [255],[0,255])
    plt.plot(hist,color=col)
```

```python
#using bitwise_and
image = cv2.imread('chill sloth.png')
blank= np.zeros(image.shape[:2], dtype=np.uint8)
blank= cv2.circle(blank,(600,500),200,(255,255,255),-1)
mask=cv2.bitwise_and(image,image,mask=blank)
mask =  cv2.cvtColor(mask,cv2.COLOR_BGR2RGB)
plt.imshow(mask)
```

```python
#Edge detection using laplacian 
image = cv2.imread('chill sloth.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
lap=cv2.Laplacian(gray,cv2.CV_64F)
lap=np.uint8(np.absolute(lap))
lap =  cv2.cvtColor(lap,cv2.COLOR_BGR2RGB)
plt.imshow(lap)
```

```python
# face detection using harr cascade

img = cv2.imread('download.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
haar_cascade = cv2.CascadeClassifier('haar_face.xml')
faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)

print(f'Number of faces found = {len(faces_rect)}')

for (x,y,w,h) in faces_rect:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), thickness=2)
img =  cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
